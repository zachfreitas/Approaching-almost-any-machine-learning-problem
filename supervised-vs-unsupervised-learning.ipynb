```python
import matplotlib.pyplot as plt  
import numpy as np  
import pandas as pd  
import seaborn as sns  
from sklearn import datasets  
from sklearn import manifold  

%matplotlib inline 
```

```python
data = datasets.fetch_openml(  'mnist_784',  version=1,  return_X_y=True, parser='auto'  )  
pixel_values, targets = data  
targets = targets.astype(int)
```

```python
single_image = pixel_values.iloc[1, :].values.reshape(28, 28)  
plt.imshow(single_image, cmap='gray')
```

```python
tsne = manifold.TSNE(n_components=2, random_state=42)  
transformed_data = tsne.fit_transform(pixel_values.iloc[:3000,:].values)  
```

```python
tsne_df = pd.DataFrame(  np.column_stack((transformed_data, targets.iloc[:3000].values)),  columns=["x", "y", "targets"]  )  

tsne_df["targets"] = tsne_df.targets.astype(int)  
```

```python
tsne_df
```

```python
grid = sns.FacetGrid(tsne_df, hue="targets")  
grid.map(plt.scatter, "x", "y").add_legend()  
```
